## Prompt

```python
"""Your task is to extract sectoral features from the given annual report of OGDCL.
Identify and extract the key features that are industry specific terminologies,
performance indicators, multiple products, and financial metrics relevant to the oil and gas sector."""
```

```python
%time
from langchain.prompts import PromptTemplate

system_prompt = """Your task is to extract relevant information from the oil and gas sector data."""
B_INST, E_INST = "<s>[INST] ", " [/INST]"
template = (
                B_INST
                + system_prompt
                + """

            Chat History: {chat_history}
            Context: {context}
            User: {question}

            """
                + E_INST
            )
prompt = PromptTemplate(input_variables=["context", "chat_history", "question"], template=template)
```

```python
%time
from langchain.prompts import PromptTemplate

# system_prompt = """You are a chatbot which helps the founders and enterpreneurs, to answer business and finance related queries.
#         If a user asks question other than business or finance, ask them to only ask question about business, entrepreneurship, and finance.
#         Given the Context and chat history answer the user questions."""

system_prompt = """You are a chatbot which helps the financial traders and analysts to asses market conditions based on technical data and news impacts.
       If a user asks question other than business or finance, ask them to only ask question about business, entrepreneurship, and finance.
       Given the Context and chat history answer the user questions."""
B_INST, E_INST = "<s>[INST] ", " [/INST]"
template = (
                B_INST
                + system_prompt
                + """

            Chat History: {chat_history}
            Context: {context}
            User: {question}

            """
                + E_INST
            )
prompt = PromptTemplate(input_variables=["context", "chat_history", "question"], template=template)
```

## Conversation Chain

```python
%time
from langchain.memory import ConversationBufferMemory
memory = ConversationBufferMemory(memory_key='chat_history', return_messages=True, output_key='answer')
```

We now initialize the ConversationalRetrievalChain:

```python
%time
from langchain.chains import ConversationalRetrievalChain

chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=db.as_retriever(),
    memory=memory,
    combine_docs_chain_kwargs={"prompt": prompt}
    )
```

```python
question = """Retreive latest news on OPEC """

result=chain(question)
print(result['answer'])
```
